```python
data = []
top_resid=[]
opp_resid=[]
top_zvalue=[]
top_area=[]
with open('output.frame1.70x70.top_areas.dat') as csvfile: 
    readCSV=csv.reader(csvfile, delimiter='\t')
    data = [r for r in readCSV]
del data[0]
del data[0]
del data[0]
for i in range(len(data)):
    top_resid.append(int(data[i][0]))
    opp_resid.append(int(data[i][1]))
    top_zvalue.append(float(data[i][2]))
    top_area.append(float(data[i][3]))
```

```python
data = []
bot_resid=[]
bot_zvalue=[]
bot_area=[]
with open('output.frame1.70x70.bottom_areas.dat') as csvfile: 
    readCSV=csv.reader(csvfile, delimiter='\t')
    data = [r for r in readCSV]
del data[0]
del data[0]
del data[0]
for i in range(len(data)):
    bot_resid.append(int(data[i][0]))
    bot_zvalue.append(float(data[i][1]))
    bot_area.append(float(data[i][2]))
```
